```python
#Overall skill demand analysis
skills = ["sql", "python", "microsoft excel", "tableau", "power bi", "aws", "azure", "spark", "hadoop", "machine learning", "deep learning", "statistics", "pytorch", "snowflake", "sas", "tensorflow", "databricks"]
# A list of common data skills to check for in the job descriptions
results = []
for skill in skills:
    count = jobs['description'].str.lower().str.contains(skill, na=False).sum()
    
    results.append({"skill": skill, "count": count})

# Counting how many job descriptions mention each of the common data skills
skills_df = pd.DataFrame(results)
skills_df= skills_df.sort_values(by='count', ascending=False)
print(skills_df)
skills_df.to_csv('../data/overall_skills_demand.csv', index=False)
# Displaying the counts of each skill in descending order
```

```python
def categorize_role(title):
    if not isinstance(title, str):
        return "Other"
    
    title = title.lower()
    if "scientist" in title:
        return "Data Scientist"
    elif "analyst" in title:
        return "Data Analyst"
    elif "engineer" in title:
        return "Data Engineer"
    else:
        return "Other"
jobs['role_category'] = jobs['position'].apply(categorize_role)
# Creating a new column to categorize job roles into Data Scientist, Data Analyst, Data Engineer,
```

```python
jobs['role_category'].value_counts()
# Checking the counts of each role category in the dataset
```

```python
# Skill Demand by Role Category
results = []
roles = ["Data Scientist", "Data Analyst", "Data Engineer"]
for role in roles:
    subset = jobs[jobs['role_category'] == role]

    for skill in skills:
        pct = subset['description'].str.lower().str.contains(skill, na=False).mean()

        results.append({"role": role, "skill": skill, "percentage": pct * 100})
skill_role_df = pd.DataFrame(results)
print(skill_role_df)
# Calculating the percentage of job descriptions that mention each skill for each role category
skill_role_df.to_csv('../data/skills_by_role.csv', index=False)
```

```python
analyst_jobs= jobs[jobs['role_category'] == "Data Analyst"]
# Creating a subset of the dataset for Data Analyst roles
analyst_jobs.value_counts('position').head(20)
# The most common Data Analyst job positions in the dataset

for analyst in analyst_jobs['position'].unique():
    print(f"Skills mentioned in {analyst} job descriptions:")
    subset = analyst_jobs[analyst_jobs['position'] == analyst]
    for skill in skills:
        pct = subset['description'].str.lower().str.contains(skill, na=False).mean()
        print(f"{skill}: {pct * 100:.2f}%")
    print("\n")
# For each unique Data Analyst job position, calculating the percentage of job descriptions that mention each skill
```
