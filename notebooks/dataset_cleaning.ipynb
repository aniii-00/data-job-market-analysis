```python
skills = ["sql", "python", "microsoft excel", "tableau", "power bi", "aws", "azure", "spark", "hadoop", "machine learning", "deep learning", "statistics", "pytorch", "snowflake", "sas", "tensorflow", "databricks"]
# A list of common data skills to check for in the job descriptions
results = []
for skill in skills:
    count = jobs['description'].str.lower().str.contains(skill, na=False).sum()
    
    results.append({"skill": skill, "count": count})

# Counting how many job descriptions mention each of the common data skills
skills_df = pd.DataFrame(results)
skills_df= skills_df.sort_values(by='count', ascending=False)
print(skills_df)
# Displaying the counts of each skill in descending order
```

```python
def categorize_role(title):
    if not isinstance(title, str):
        return "Other"
    
    title = title.lower()
    if "scientist" in title:
        return "Data Scientist"
    elif "analyst" in title:
        return "Data Analyst"
    elif "engineer" in title:
        return "Data Engineer"
    else:
        return "Other"
jobs['role_category'] = jobs['position'].apply(categorize_role)
# Creating a new column to categorize job roles into Data Scientist, Data Analyst, Data Engineer,
```

```python
jobs['role_category'].value_counts()
# Checking the counts of each role category in the dataset
```
